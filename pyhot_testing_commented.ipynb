Try to check out the SLM calculating code. Here are the relevant physical parameters:

Here's a plausibility check. For a trap at position $X$, the phase variation across the entire sensor ($x=1200\times 8$ micrometers) should be given by

$$\Delta \phi = \frac{2\pi x X}{\lambda f}.$$

The number of times the pattern should repeat should then be given by $$\Delta \phi / (2\pi)$$.

```python
2*np.pi*1200*8*2/(1.064*3333.33)
```

```python
_/(2*np.pi)
```

The pattern above does in fact repeat a bit over 5 times.
